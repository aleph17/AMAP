```python
import numpy as np
import os
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
mask_dir = '/home/muhammad-ali/github/AMAP/segmentation/masks'

mask_list = os.listdir(mask_dir)
mask_dirs = [os.path.join(mask_dir, x) for x in mask_list]

distribution = np.zeros([256], np.uint64)
for i in range(len(mask_dirs)):
    mask = cv.imread(mask_dirs[i])
    dimension_values = np.array(mask)
    temp = np.bincount(mask.reshape(1,-1)[0])
    distribution += temp.astype(np.uint64)
# print(distribution)

indices = range(len(distribution))  # Indices represent each unique value
plt.bar(indices, distribution, color='lightgreen', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Frequency Distribution by Value")
plt.show()
```
